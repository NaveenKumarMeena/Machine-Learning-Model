# WithOut Gredient decent --- Normal function

```python
import numpy as np
#taking features size,floors,bedrooms
x=np.array([[1,100,2,2],[1,150,1,3],[1,125,2,3]])
y=np.array([15,13,18])

y=y.reshape((3,1))
t0,t1,t2,t3=0.3,1,0.1,0.05

#theta values
tt=np.array([t0,t1,t2,t3])

xtheta=np.multiply(tt,x)

rss=(np.sum(xtheta-y)**2)
print("error",rss)

xT = np.transpose(x)
xT


b = np.dot(xT,y)
a = np.linalg.inv(np.dot(xT,x))
theta_new = np.dot(a,b)
theta = np.array(theta_new)
theta
```

```python
y_hat = np.dot(x,theta)
error = np.mean((y_hat - y) ** 2)
error
```
