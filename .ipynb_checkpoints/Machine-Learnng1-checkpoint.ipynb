```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Size = [100 , 200 , 300 , 250 , 150]
Price = [5,10,20,18,8]
plt.scatter(Size , Price, color='blue' )
plt.show()
```

```python
theta0 = 1
theta1 = 1

yhat1 = (theta0 + theta1*Size[0])
yhat2 = (theta0 + (theta1*Size[1]))
yhat3 = (theta0 + (theta1*Size[2]))
yhat4 = (theta0 + (theta1*Size[3]))
yhat5 = (theta0 + (theta1*Size[4]))
yhat = [yhat1, yhat2, yhat3 , yhat4, yhat5]

# Mean Square Error

Error1 = yhat1 - Price[0]
Error2 = yhat2 - Price[1]
Error3 = yhat3 - Price[2]
Error4 = yhat4 - Price[3]
Error5 = yhat5 - Price[4]
print("Error: ",Error5)
```

```python
Error = (Error1**2) + (Error2**2) + (Error3**2) + (Error4**2) + (Error5**2)
Errormean = 1/5* Error
Errormean
```

```python
# Plot theta= 0 , theta1 = 1 (intercept = 0  , slope = 1)
theta0 = 0
theta1 = 1
yhat1 = (theta0 + (theta1*Size[0]))
yhat2 = (theta0 + (theta1*Size[1]))
yhat3 = (theta0 + (theta1*Size[2]))
yhat4 = (theta0 + (theta1*Size[3]))
yhat5 = (theta0 + (theta1*Size[4]))
yhat = [yhat1, yhat2, yhat3 , yhat4 , yhat5]
yhatfirst = yhat;


plt.figure(figsize = (10,10))
plt.grid(color='k', linestyle='-', linewidth=0.5)
plt.scatter(Size , Price, color='blue' )
plt.plot(Size, yhat, '-r')
plt.xlabel("size")
plt.ylabel("price")
print("yhat : " , yhat)
```
