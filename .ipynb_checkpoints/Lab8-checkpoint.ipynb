```python
import math
data = [10,20,30,40,41,50,60,70,80,90,100,120,130,140]
meu1 = 15
sigma1 = 2
meu2 = 95
sigma2 = 3
for j in range(25):
    k1=0
    k2=0
    l1=[]
    l2=[]
    c1=[]
    c2=[]
    for i in range(len(data)):
        flag1 = False
        flag2=False
        if sigma1>0:
            l1.append(((2*math.pi)**(1/2)*sigma1)**(-1)* math.e**(((-1)*(data[i]-meu1)**2)*(2*sigma1*sigma1)**(-1)))
            flag1=True
        if sigma2>0:
            l2.append(((2*math.pi)**(1/2)*sigma2)**(-1)* math.e**(((-1)*(data[i]-meu2)**2)*(2*sigma2*sigma2)**(-1)))
            flag2=True
        if  (flag1 and flag2) and l1[k1]>l2[k2]:
            c1.append(data[i])
            l1.pop()
            l2.pop()
            flag1 = False
            flag2=False
        else :
            c2.append(data[i])
            l1.pop()
            l2.pop()
            flag1 = False
            flag2=False
        
        if flag1:
           c1.append(data[i]) 
           l1.pop()
        
        if flag2:
            c2.append(data[i])
            l2.pop()

    if len(c1)>0 :
        meu1 = sum(c1)/len(c1)
        y=0
        for i in range(len(c1)):
            y += (c1[i]-meu1)**2
        
        sigma1 = (y/len(c1))**(1/2)
        
    if len(c2)>0 :
        meu2 = sum(c2)/len(c2)
        y=0
        for i in range(len(c2)):
            y += (c2[i]-meu2)**2

        sigma2 = (y/len(c2))**(1/2)

print(c1)
print(c2)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('Clustering_gmm.csv')
plt.figure(figsize=(7,7))
plt.scatter(data["Weight"],data["Height"])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Data Distribution')
plt.show()
```

```python
#training k-means model
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(data)
#predictions from kmeans
pred = kmeans.predict(data)
frame = pd.DataFrame(data)
frame['cluster'] = pred
frame.columns = ['Weight', 'Height', 'cluster']
#plotting results
color=['blue','green','cyan', 'black']
for k in range(0,4):
    data = frame[frame["cluster"]==k]
    plt.scatter(data["Weight"],data["Height"],c=color[k])
plt.show()
```

```python
import pandas as pd
data = pd.read_csv('Clustering_gmm.csv')
# training gaussian mixture model
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=4)
gmm.fit(data)
#predictions from gmm
labels = gmm.predict(data)
frame = pd.DataFrame(data)
frame['cluster'] = labels
frame.columns = ['Weight', 'Height', 'cluster']
color=['blue','green','cyan', 'black']
for k in range(0,4):
    data = frame[frame["cluster"]==k]
    plt.scatter(data["Weight"],data["Height"],c=color[k])
plt.show()
```
