```python
import matplotlib.pyplot as plt
import numpy as np


def gd(alpha,n):
    x=np.array([100,150,300,200,250])
    y=np.array([5,8,20,15,18])
    theta1=0
    theta0=0
    m=len(x)
    J=[]
    for i in range(n):
        y_cap=theta1*x+theta0
        error=y_cap-y
        theta1=theta1-alpha*(1/m)*np.sum(error*x)
        theta0=theta0-alpha*(1/m)*np.sum(error)
        J.append((1/(2*m))*np.sum(error**2))     
    
    

    plt.plot(range(n),J)
    plt.xlabel("Number of iterations")
    plt.ylabel("Cost function, J")
    plt.show()

print("Case 1: alpha=0.0001")
gd(0.00001,50)
print("Case 2: alpha=0.001")
gd(0.001,50)
print("Case 3: alpha=000001",)
gd(0.000001,100)
```
