```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# define the data points
xx1 = [[185,170,168,179,182,188,180,180,183,180,180,170]]
xx2 = [[72,56,60,68,72,77,71,80,84,88,67,76]]
x1 = np.array(xx1)
x2 = np.array(xx2)

plt.scatter(x1[0],x2[0])
plt.show()

# set the initial centroids
c1 = [x1[0][6], x2[0][6]]
c2 = [x1[0][1], x2[0][1]]
print(c1)
print(c2)


flag =True
while(flag):
    # calculate the Euclidean distance between each data point and the centroids
    distancesc1 = np.sqrt(((x1 - c1[0])**2) + ((x2 - c1[1])**2))
    distancesc2 = np.sqrt(((x1 - c2[0])**2) + ((x2 - c2[1])**2))
    
    labels = np.where(distancesc1 < distancesc2, 1, 2)
    
    # update centroids based on mean of labeled points
    new_c1 = [np.mean(x1[labels == 1]), np.mean(x2[labels == 1])]
    new_c2 = [np.mean(x1[labels == 2]), np.mean(x2[labels == 2])]
   
    if (new_c1 == c1) and (new_c2 == c2):
        flag = False
    else:
        c1 = new_c1
        c2 = new_c2


    # print the centroids and labels for each iteration
    print(f"Iteration {i+1}:")
    print(f"Centroid 1: {c1}")
    print(f"Centroid 2: {c2}")
    print(f"Labels: {labels}")
    print()
# plot the data points and centroids
plt.scatter(x1[0], x2[0], c=labels)
plt.scatter(c1[0], c1[1], marker='*', s=200)
plt.scatter(c2[0], c2[1], marker='*', s=200)
plt.show()
```

```python
data = pd.read_csv('Cluster_dummy.csv')
plt.figure(figsize=(7,7))
plt.scatter(data["Weight"],data["Height"])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Data Distribution')
plt.show()
```

```python
#training k-means model
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)
#predictions from kmeans
pred = kmeans.predict(data)
frame = pd.DataFrame(data)
frame['cluster'] = pred
frame.columns = ['Weight', 'Height', 'cluster']
#plotting results
color=['blue','green']
for k in range(0,2):
    data = frame[frame["cluster"]==k]
    plt.scatter(data["Weight"],data["Height"],c=color[k])
plt.show()
```
